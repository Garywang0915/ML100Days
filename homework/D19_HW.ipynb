```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)
```

```python
# 開始畫圖
plt.plot(x,y)
# 在這個指令之前，都還在做畫圖的動作 
# 這個指令算是 "秀圖"
plt.show() 
```

```python
#畫出完整的 sin 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)

plt.plot(x, y_sin)

plt.show()
```

```python
#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)

plt.plot(x, y_cos)

plt.savefig("filename.png",dpi=300,format="png")
```

```python
#顏色由（X，Y）的角度給出。

#注意標記的大小，顏色和透明度。

n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

Angle = np.degrees(np.arctan2(Y,X)) # 計算夾角(單位degree)

plt.scatter(X, Y, c = Angle, alpha = 0.5, s = 100)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
